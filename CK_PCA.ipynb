```python
import ipyrad.analysis as ipa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = r"CK_ipyrad_outfiles/test.snps.hdf5"
```

```python
# for i in pca.names:
#     print(f"'{i}':['{i}'],")
```

```python
imap = {'CKpop1':['CK18','CK11','CK19'],
'CKpop2':['CK16', 'CK14', 'CK20', 'CK13', 'CK17', 'CK23'],
'CKpop3':['CK22','CK21','CK25','CK12','CK24','CK10','CK01','CK08'],
'CKpop4':['CK15','CK09'],
'CKpop5':['CK05','CK04','CK02','CK03','CK07','CK06'],
}

minmap = {i:0.5 for i in imap}
```

```python
pca =ipa.pca(data,
             imap=imap,
            impute_method= "sample",# deal with the missing data, sample =randomly sample genotype from the data)
             mincov=0.8, # give the minimum coverage threshold
             minmap=minmap,
             
            ) 
```

```python
# pca.run_tsne(perplexity=5.0, n_iter=10000, verbose=True) # seed is for get the repeatable results
pca.run(seed=3333)
```

```python

pca.draw(0,1)
```
